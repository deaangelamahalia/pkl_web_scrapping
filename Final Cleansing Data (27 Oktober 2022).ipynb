# KOPI

```python
import requests #Mengimport library request untuk memanggil url
import json #Mengimport library json untuk mengubah data dictionary menjadi teks json
import pandas as pd #Mengimport library pandas untuk mengubah file json menjadi dataframe
import re #Mengimport library re/regex untuk pencarian string atau teks dengan menggunakan pola (pattern)
```

```python
kopi = pd.read_csv("kopi.csv") # Membaca data pada kopi.csv
```

```python
kopi
```

```python
for page in kopi['Nama']: #Melakukan looping untuk menampilkan data berdasarkan kolom nama
    print(page)
```

```python
# Membuat List
kemasan = ["Sachet", "Botol", "Kaleng"]
satuan = ["ml", "g"]
hasil = []

# Melakukan Fungsi Filtering/Substring pada kolom nama menjadi kolom sub baru 
for i in kopi['Nama']: # Melakukan looping untuk menampilkan nama
    if any(daftar in i for daftar in kemasan):
        for a in kemasan:
            if not i.find(a) == -1:
                wadah = a
                break
        nama = i[:i.find(wadah)]
        berat = i[i.find(wadah)+(len(wadah)+1):]
        hasil.append([nama, wadah, berat])
    else:
        bobot = int(re.search(r'\d+', i).group())
        berat = i[i.find(str(bobot)):]
        nama = i[:i.find(str(bobot))]
        temp = i.replace(f"{nama}","")
        for a in satuan:
            if not temp.find(a) == -1:
                if a == "ml":
                    wadah = "Botol/Kaleng"
                elif a == "g":
                    wadah = "Sachet"
        hasil.append([nama, wadah, berat])
        
for i in hasil:
    print(f"Nama Produk : {i[0]} \SWadah Produk : {i[1]}\sBerat Produk : {i[2]}")
```

```python
data = {}
data['nama']  = []
data['kemasan'] = []
data['berat'] = []

for i in hasil:
    data['nama'].append(i[0])
    data['kemasan'].append(i[1])
    data['berat'].append(i[2])
```

```python
data
```

```python
kopi["Kemasan"] = data["kemasan"]
kopi["Merk"] = data["nama"]
kopi["Berat"] = data["berat"]
```

```python
kopi
```

```python
kopi=kopi[["Nama","Merk","Kemasan","Berat","Harga"]]
```

```python
brand = []
for i in kopi["Merk"]:
    kata=i.split(' ')
    produk = kata[:len(kata)//2]
    produkx = ' '.join(produk)
    brand.append(produkx)
```

```python
brand
```

```python
kopi["Merk"] = brand
```

```python
kopi["Merk"]
```

```python
kopi
```

```python
kopi.to_csv('Produk Kopi (Final) 1.csv', index=False)
```

```python
for page in minyak['Nama']:
    print(page)
```

```python
kemasan = ["Pouch", "Botol", "Cup"]
satuan = ["ml", "L"]
hasil = []

for i in minyak['Nama']:
    if any(daftar in i for daftar in kemasan):
        for a in kemasan:
            if not i.find(a) == -1:
                wadah = a
                break
        nama = i[:i.find(wadah)]
        berat = i[i.find(wadah)+(len(wadah)+1):]
        hasil.append([nama, wadah, berat])
    else:
        bobot = int(re.search(r'\d+', i).group())
        berat = i[i.find(str(bobot)):]
        nama = i[:i.find(str(bobot))]
        temp = i.replace(f"{nama}","")
        for a in satuan:
            if not temp.find(a) == -1:
                if a == "ml":
                    wadah = "Pouch/Cup"
                elif a == "L":
                    wadah = "Botol"
        hasil.append([nama, wadah, berat])
        
```

```python
data_m = {}
data_m['nama']  = []
data_m['kemasan'] = []
data_m['berat'] = []

for i in hasil:
    data_m['nama'].append(i[0])
    data_m['kemasan'].append(i[1])
    data_m['berat'].append(i[2])
```

```python
data_m["nama"]
```

```python
minyak["Kemasan"] = data_m["kemasan"]
minyak["Merk"] = data_m["nama"]
minyak["Berat"] = data_m["berat"]
```

# KECANTIKAN

```python
cantik = pd.read_csv("kecantikan.csv")
```

```python
cantik
```

```python
for page in cantik['Nama']:
    print(page)
```

```python
list_angka = [1, 2, 3, 4, 5, 6, 7, 8, 9, 0]
list_angka = [str(x) for x in list_angka]
list_berat = []
list_nama = []
list_kata = []
list_kata2 = []
list_angka
```

```python
for page in cantik['Nama']:
    start_idx = -1 # index untuk angka
    for idx in range(0, len(page)):
        # mencari index dari sebuah char yg termasuk angka 
        if(page[idx] in list_angka):
            start_idx = idx
            break
    
    berat = page[start_idx:len(page)]
    list_berat.append(berat)
    page = page[0:start_idx-1]
    list_nama.append(page)
    
    kata=page.split()
    pecahan = kata[0]
    list_kata2.append(pecahan)
    print(pecahan)
```

```python
cantik["Kemasan"]= list_berat
cantik["Merk"] = list_nama
cantik["Brand"] = list_kata2
```

```python
cantik
```

```python
cantik = cantik[["Nama","Brand","Kemasan","Harga"]]
```
